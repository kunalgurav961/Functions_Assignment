# Theory Questions and Answers

### 1. What is the difference between a function and a method in Python?

A **function** is a block of code that is defined using `def` and can be called independently. A **method** is similar but is associated with an object (i.e., it's a function that belongs to a class).
     
Example:
```python
def greet(): return "Hello"          # Function
print("hello".upper())               # upper() is a method
```

### 2. Explain the concept of function arguments and parameters in Python.

**Parameters** are variables listed in a function definition. **Arguments** are the values passed to the function.
     
Example:
```python
def add(a, b):     # a, b are parameters
    return a + b

print(add(3, 4))   # 3, 4 are arguments
```

### 3. What are the different ways to define and call a function in Python?

Functions can be defined using `def` or `lambda`. You call them by their name followed by parentheses.

Example:
```python
def greet(name):
    return f"Hello {name}"

print(greet("Alice"))
```

### 4. What is the purpose of the `return` statement in a Python function?

The `return` statement sends a result back to the caller and exits the function.

Example:
```python
def square(x):
    return x * x
```

### 5. What are iterators in Python and how do they differ from iterables?

An **iterable** is any object you can loop over (e.g., list, string). An **iterator** is an object with a `__next__()` method.

Example:
```python
lst = [1, 2, 3]          # iterable
it = iter(lst)           # iterator
print(next(it))          # prints 1
```

### 6. Explain the concept of generators in Python and how they are defined.

Generators are functions that yield values one at a time using the `yield` keyword.

Example:
```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```

### 8. What is a lambda function in Python and when is it typically used?

A **lambda** is an anonymous one-liner function used for simple tasks.

Example:
```python
square = lambda x: x * x
print(square(5))
```

### 9. Explain the purpose and usage of the `map()` function in Python.

`map()` applies a function to every item of an iterable.

Example:
```python
nums = [1, 2, 3]
squares = list(map(lambda x: x**2, nums))
```

### 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- `map()`: Applies function to each element.
- `reduce()`: Aggregates the elements to a single value.
- `filter()`: Selects elements for which the function returns `True`.

Example:
```python
from functools import reduce
nums = [1, 2, 3, 4]
print(list(map(lambda x: x*2, nums)))
print(reduce(lambda x, y: x+y, nums))
print(list(filter(lambda x: x%2 == 0, nums)))
```

# Practical Questions and Answers

### 1. Sum of all even numbers in a list

```python
def sum_even(lst):
    return sum(x for x in lst if x % 2 == 0)
```

### 2. Reverse a string

```python
def reverse_string(s):
    return s[::-1]
```

### 3. Square each integer in a list

```python
def square_list(lst):
    return [x**2 for x in lst]
```

### 4. Prime checker from 1 to 200

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

### 5. Iterator class for Fibonacci sequence

```python
class Fib:
    def __init__(self, limit):
        self.a, self.b = 0, 1
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.a > self.limit:
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        return val
```

### 6. Generator yielding powers of 2 up to a given exponent

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i
```

### 7. Generator yielding lines from a string

```python
def line_reader(text):
    for line in text.splitlines():
        yield line
```

### 8. Lambda to sort by second element

```python
lst = [(1, 3), (2, 2), (3, 1)]
lst.sort(key=lambda x: x[1])
```

### 9. Convert Celsius to Fahrenheit using map

```python
celsius = [0, 10, 20]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
```

### 10. Remove vowels from a string using filter

```python
s = "Hello World"
no_vowels = ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
```
